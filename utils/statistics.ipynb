```python
import matplotlib.pyplot as plt

dates = ['2023.02.20', '2023.02.21', '2023.02.22', '2023.02.23', '2023.03.02', '2023.03.03', '2023.03.06', '2023.03.07', '2023.03.09', '2023.03.10', '2023.03.13', '2023.03.17', '2023.03.20', '2023.03.21', '2023.03.23', '2023.03.24', '2023.03.28', '2023.03.29', '2023.03.30', '2023.03.31']
problems_solved = [3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 4, 4, 4]

plt.plot(dates, problems_solved, marker='o')
plt.xticks(rotation=45)
plt.ylim((0, max(problems_solved)+1))
plt.xlabel('Date')
plt.ylabel('Number of Problems Solved')
plt.title('Number of Problems Solved from Feb to Mar')
plt.show()
```
